# **Feature Engineering**

```python
import pandas as pd
import numpy as np
```

```python
#loading dataset
from pathlib import Path
PROJECT_ROOT = Path.cwd().parent
DATA_PATH = PROJECT_ROOT / "data" / "processed" / "cleaned_upi_fraud_dataset.csv"
assert DATA_PATH.exists(), f"Dataset not found at {DATA_PATH}"

df = pd.read_csv(DATA_PATH, encoding="utf-8")
```

```python
df_fe = df.copy()

if pd.api.types.is_categorical_dtype(df_fe['hour']):
    df_fe['hour'] = df_fe['hour'].astype(int)

if pd.api.types.is_categorical_dtype(df_fe['day_of_week']):
    df_fe['day_of_week'] = df_fe['day_of_week'].astype(int)

elif df_fe['day_of_week'].dtype == object:
    days_map = {'Monday': 0, 'Tuesday': 1, 'Wednesday': 2, 'Thursday': 3,
                'Friday': 4, 'Saturday': 5, 'Sunday': 6}
    df_fe['day_of_week'] = df_fe['day_of_week'].map(days_map).astype(int)

df_fe['hour_sin'] = np.sin(2 * np.pi * df_fe['hour'] / 24)

df_fe['hour_cos'] = np.cos(2 * np.pi * df_fe['hour'] / 24)

df_fe['day_of_week_sin'] = np.sin(2 * np.pi * df_fe['day_of_week'] / 7)

df_fe['day_of_week_cos'] = np.cos(2 * np.pi * df_fe['day_of_week'] / 7)

df_fe['amount_log'] = np.log1p(df_fe['amount (INR)'])

drop_cols = ['timestamp', 'date', 'hour', 'day_of_week', 'amount (INR)', 'transaction id']

df_fe = df_fe.drop(columns=[c for c in drop_cols if c in df_fe.columns])
```

```python
df_fe.to_csv(PROJECT_ROOT / "data" / "processed" / "upi_features.csv", index=False)
```

## Feature Engineering Rationale
- Time-based features capture behavioral patterns
- Device and network features capture fraud signals
- Aggregations help identify abnormal behavior
